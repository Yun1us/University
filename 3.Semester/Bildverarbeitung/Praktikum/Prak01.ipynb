```python
import numpy as np
from PIL import Image

# Load the image
img = Image.open("Moonius.jpg")
img_array = np.array(img)

# Define the box filter kernel
kernel_size = 9  # Adjust the kernel size as needed
kernel = np.ones((kernel_size, kernel_size)) / (kernel_size * kernel_size)

# Apply the filter to each color channel
filtered_array = np.zeros_like(img_array)
for c in range(3):  # Iterate over RGB channels
    for i in range(kernel_size // 2, img_array.shape[0] - kernel_size // 2):
        for j in range(kernel_size // 2, img_array.shape[1] - kernel_size // 2):
            filtered_array[i, j, c] = np.sum(img_array[i - kernel_size // 2:i + kernel_size // 2 + 1, j - kernel_size // 2:j + kernel_size // 2 + 1, c] * kernel)

# Convert the filtered array back to an image
filtered_img = Image.fromarray(filtered_array.astype(np.uint8))

# Display or save the filtered image
filtered_img.show()  # To display
#Tosave
```
