```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(10,6), dpi=100)

# Plot sine wave
plt.plot(x, y1, 'b-', label='Sine wave')

# Plot cosine wave
plt.plot(x, y2, 'r--', label='Cosine wave')

# Add a title (specify font parameters with fontdict)
plt.title('Sine and Cosine Waves', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X values')
plt.ylabel('Y values')

# X, Y axis Ticks (scale of your graph)
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(-1, 1.5, 0.5))

# Add a Legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('sine_cosine_plot.png', dpi=300)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.array([0, 1, 2, 3, 4, 5])
y = 2 * x + 1

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Plot the linear function
plt.plot(x, y, 'bo-', label='y = 2x + 1')

# Add a title (specify font parameters with fontdict)
plt.title('Linear Function: y = 2x + 1', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X values')
plt.ylabel('Y values')

# X, Y axis Ticks (scale of your graph)
plt.xticks(np.arange(0, 6, 1))
plt.yticks(np.arange(0, 12, 2))

# Add a Legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('linear_function_plot.png', dpi=300)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Data
np.random.seed(10)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1.5, 100)
data3 = np.random.normal(2, 0.5, 100)

# Combine data into a single dataset
data = [data1, data2, data3]

# Create labels for the data
labels = ['Group 1', 'Group 2', 'Group 3']

# Plot the violin plot
plt.figure(figsize=(8, 5), dpi=100)
sns.violinplot(data=data)

# Add labels and title
plt.xticks([0, 1, 2], labels)
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Violin Plot of Different Groups')
plt.savefig('violin_plot.png', dpi=300)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Setting Seaborn style
sns.set()

# Load the Titanic dataset
df = pd.read_csv("./titanic.csv")

# Prepare the data: Drop NaN values and convert to integers
data = df["Age"].dropna().astype(int)

# Plot the normalized histogram
counts, bins, patches = plt.hist(data, bins=16, color="skyblue", edgecolor='black', density=True)

# Set plot labels and title
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Normalized Age Distribution")

# Show the plot
plt.show()

# Verify the sum of the heights of the rectangles equals 1
bin_width = np.diff(bins)
heights = counts * bin_width
sum_of_heights = np.sum(heights)

print(f"Sum of the heights of the rectangles: {sum_of_heights}")

```
