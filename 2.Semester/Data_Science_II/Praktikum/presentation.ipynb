```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
import matplotlib.image as mpimg

gewichtungen = {
    'CDU/CSU': 2,
    'SPD': 1,
    'GRÜNE': 0.5,
    'FDP': 1.5,
    'DIELINKE': 0.5,
    'AfD': 4
}

daten = pd.read_csv('./hdata2013.csv')

daten['Gesamtgewichtung'] = (
    (daten['CDU/CSU'] * gewichtungen['CDU/CSU'] +
     daten['SPD'] * gewichtungen['SPD'] +
     daten['GRÜNE'] * gewichtungen['GRÜNE'] +
     daten['FDP'] * gewichtungen['FDP'] +
     daten['DIELINKE'] * gewichtungen['DIELINKE'] +
     daten['AfD'] * gewichtungen['AfD']) / daten['GueltigeWaelende']
)

norm_gesamtgewichtung = plt.Normalize(daten['Gesamtgewichtung'].min(), daten['Gesamtgewichtung'].max())

colormap_blau_orange = plt.cm.coolwarm

gesamtgewichtung_farben = colormap_blau_orange(norm_gesamtgewichtung(daten['Gesamtgewichtung']))

blasen_groessen = daten['GueltigeWaelende'] / 15000 

koordinaten = {
    'Schleswig-Holstein': (6.5, 15.5),
    'Hamburg': (7, 14),
    'Bremen': (5.25, 13),
    'Niedersachsen': (5.5, 12),
    'Mecklenburg-Vorpommern': (9.5, 14),
    'Nordrhein-Westfalen': (5, 8.5),
    'Rheinland-Pfalz': (3.5, 6),
    'Saarland': (3.0, 5),
    'Hessen': (5.5, 5),
    'Berlin': (10, 12),
    'Brandenburg': (10.75, 10.5),
    'Sachsen': (10, 8.5),
    'Sachsen-Anhalt': (8, 10.5),
    'Thüringen': (8, 7),
    'Bayern': (8.5, 3.5),
    'Baden-Württemberg': (6, 3.0)
}

daten['x'] = daten['Bundesland'].map(lambda bl: koordinaten[bl][0])
daten['y'] = daten['Bundesland'].map(lambda bl: koordinaten[bl][1])

hintergrund_bild = mpimg.imread('./image.png')

plt.figure(figsize=(12, 12))

plt.imshow(hintergrund_bild, extent=[0, 14, 0, 17])

scatter = plt.scatter(daten['x'], daten['y'], s=blasen_groessen*25, c=gesamtgewichtung_farben, alpha=0.6, edgecolors='w')

plt.xlim(1, max(daten['x']) + 3)
plt.ylim(0, max(daten['y']) + 1.5)

plt.gca().set_xticks([])
plt.gca().set_yticks([])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

for i, row in daten.iterrows():
    plt.text(row['x'], row['y'], row['Bundesland'], horizontalalignment='center', verticalalignment='center')

legende_elemente = [
    plt.Line2D([0], [0], marker='o', color='w', label='Links (Blau)',
               markerfacecolor='blue', markersize=10),
    plt.Line2D([0], [0], marker='o', color='w', label='Rechts (Rot)',
               markerfacecolor='red', markersize=10)
]
plt.legend(handles=legende_elemente, loc='upper left', title='Politische Ausrichtung')

plt.title('Politische Gewichtung der Bundesländer 2013')

plt.show()
```
